# Métodos hierárquicos de agrupamento - Tarefa

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler

import scipy.cluster.hierarchy as shc

from IPython.display import display
```

## 1) Repetindo o que fizemos em aula

Primeiramente vamos repetir o que fizemos em aula com a base de pinguins.

1. Selecione as variáveis quantitativas da base de pinguins
2. Elimine valores faltantes
3. Padronize as variáveis para média zero e desvio padrão 1
4. Faça um agrupamento hierárquico dividindo essa base em 3 grupos
5. Faça o Dendrograma

### 1. Selecione as variáveis quantitativas da base de pinguins

```python
base_peng = sns.load_dataset('penguins')
base_peng.index.name='id'
print("Amostra da base de dados de Pinguins:")
display(base_peng.sample(5, random_state=0))
print("Informações adicionais sobre a base de dados:", end="\n\n")
print(base_peng.info())
```

Podemos ver que asvariáves quantitativas são: `bill_length_mm`, `bill_depth_mm`, `flipper_length_mm` e `body_mass_g`.

```python
base_quanti = base_peng.select_dtypes(include='number')
display(base_quanti.head())
```

### 2. Elimine valores faltantes

Vamos checar a quantidade de valores nulos em cada coluna quantitativa:

```python
base_quanti.isnull().sum()
```

```python
print(f"Tamanho original: {len(base_quanti)}")
print(f"Tamanho sem valores faltantes: {len(base_quanti.dropna())}")
```

São duas instâncias inteiras que apresentam valores faltantes em todas as colunas quantitativas.

```python
base_s_falt = base_quanti.dropna()
```

### 3. Padronize as variáveis para média zero e desvio padrão 1

```python
standard_scaler = StandardScaler()
base_padron = standard_scaler.fit_transform(base_s_falt)
base_padron
```

Podemos ver se estão realmente padronizadas:

```python
print("Média:", base_padron.mean())
print("Desvio Padrão:",base_padron.std())
```

### 4. Faça um agrupamento hierárquico dividindo essa base em 3 grupos

Para construir os labels na base de dados, fazemos:

```python
# complete
hier_clustering = AgglomerativeClustering(linkage="complete", 
                                           distance_threshold = None, 
                                           n_clusters=3)
hier_clustering.fit(base_padron)
base_i = base_s_falt.copy(deep=False)
base_i["grupo_complete"] = hier_clustering.labels_

# average
hier_clustering = AgglomerativeClustering(linkage="average", 
                                           distance_threshold = None, 
                                           n_clusters=3)
hier_clustering.fit(base_padron)
base_i["grupo_average"] = hier_clustering.labels_

# single
hier_clustering = AgglomerativeClustering(linkage="single", 
                                           distance_threshold = None, 
                                           n_clusters=3)
hier_clustering.fit(base_padron)
base_i["grupo_single"] = hier_clustering.labels_

# juntando tudo!
base_final = base_peng.merge(base_i[['grupo_single', 'grupo_complete', 'grupo_average']], how='left', on='id')
base_final
```

### 5. Faça o Dendrograma

Vamos comparar 3 tipos de linkage: `single` (vizinho mais próximo), `complete` (vizinho mais distante) e `average` (distância média).

Para isso vamos ter que utilizar a base com valores quantitativos e sem faltantes.

```python
dend_single = shc.dendrogram(shc.linkage(base_s_falt, method='single'))
```

```python
dend_complete = shc.dendrogram(shc.linkage(base_s_falt, method='complete'))
```

```python
dend_average = shc.dendrogram(shc.linkage(base_s_falt, method='average'))
```

Podemos ver que o método `single` parece não trazer muita informação para análise, enquanto `complete` e `average` conseguem dividir grupos distintos com melhor visualização. Sabe-se que o método `single` não funciona bem para dados que são mais próximos/homogêneos entre si.

Vamos utilizar o Método Elbow para encontrar o melhor número de agrupamentos para essa base de dados, de forma que vemos qual 'ganho' nós temos ao aumentar ou diminuir o número de agrupamentos.

```python
from yellowbrick.cluster import KElbowVisualizer

model = AgglomerativeClustering()
visualizer = KElbowVisualizer(model, k=(4,12))

visualizer.fit(base_padron)
visualizer.show()  
```

O melhor número de agrupamentos parece ser **5**, vamos usar o método **complete**.

```python
hier_clustering = AgglomerativeClustering(linkage="complete", 
                                           distance_threshold = None, 
                                           n_clusters=5)
hier_clustering.fit(base_padron)
base_i = base_s_falt.copy(deep=False)
base_i["grupo_complete"] = hier_clustering.labels_
base_i
```
